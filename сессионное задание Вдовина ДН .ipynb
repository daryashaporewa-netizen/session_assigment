(a). $y=e^{x\ln(3x+1)}$.

**Область определения.** Требуем $3x+1>0 \Rightarrow x>-1/3$.

Положим $g(x)=x\ln(3x+1)$. Тогда $y=e^{g(x)}$, и по правилу производной экспоненты
$$
y' = e^{g(x)}\cdot g'(x).
$$

Найдём $g'(x)$ (правило произведения):
$$
g'(x)=(x)'\ln(3x+1)+x\cdot\frac{(3x+1)'}{3x+1}
=\ln(3x+1)+\frac{3x}{3x+1}.
$$

**Итог:**
$$
\boxed{\,y' = e^{x\ln(3x+1)}\!\left(\ln(3x+1)+\frac{3x}{3x+1}\right),\quad x>-1/3\,}
$$

$$
x^3+4x^2-7\;\ge 0.
$$

Положим $u(x)=e^{x^2}$ и $v(x)=\sqrt{x^3+4x^2-7}=(x^3+4x^2-7)^{1/2}$. Тогда $y=u\cdot v$, и по правилу произведения
$$
y' = u'\,v + u\,v'.
$$

Найдём производные $u'$ и $v'$ (правила цепочки и степени):
$$
u'(x)=\frac{d}{dx}\,e^{x^2}=2x\,e^{x^2},\qquad

$$
\boxed{\;

(Формула производной корректна при \(|x|<1\).)

Обозначим
$$
N(x)=x\sqrt{1-x^2},\qquad D(x)=1+x^2,\qquad y=\frac{N}{D}.
$$
Тогда по правилу частного
$$
y'=\frac{N'D-ND'}{D^{2}}.
$$

**Вычислим \(N'(x)\). Пусть**
$$
s(x)=\sqrt{1-x^2}=(1-x^2)^{1/2}.
$$
$$
s'(x)=\tfrac{1}{2}(1-x^2)^{-1/2}\cdot(-2x)=-\frac{x}{\sqrt{1-x^2}}.
$$
Отсюда
$$
N'(x)=1\cdot s(x)+x\cdot s'(x)
=\sqrt{1-x^2}-\frac{x^2}{\sqrt{1-x^2}}
=\frac{1-2x^2}{\sqrt{1-x^2}},\qquad
D'(x)=2x.
$$

Подставляя в формулу частного:
$$
y'=\frac{\dfrac{1-2x^2}{\sqrt{1-x^2}}\,(1+x^2)\;-\;x\sqrt{1-x^2}\cdot(2x)}{(1+x^2)^2}.
$$

Приведём к общему знаменателю и упростим:
$$
y'=\frac{(1-2x^2)(1+x^2)-2x^2(1-x^2)}{(1+x^2)^2\sqrt{1-x^2}}
=\frac{1-3x^2}{(1+x^2)^2\sqrt{1-x^2}}.
$$

**Итог**
$$
\boxed{\,y'=\dfrac{1-3x^2}{(1+x^2)^2\sqrt{1-x^2}}\,},\qquad |x|<1.

### Задание 3.
### Найдите условный экстремум функции, если задано условие:

Функция и ограничение

$$
z = e^{-(x+y)}, \qquad \frac{x^2}{4} + y^2 = 5 .
$$

---

### Метод множителей Лагранжа

Лагранжиан

$$
\mathcal{L}(x,y,\lambda) \;=\; e^{-(x+y)} \;+\; \lambda\!\left(\frac{x^2}{4}+y^2-5\right).
$$

Условия первого порядка

$$
\begin{cases}
\mathcal{L}_x = -\,e^{-(x+y)} + \lambda\,\dfrac{x}{2} = 0,\\[6pt]
\mathcal{L}_y = -\,e^{-(x+y)} + 2\lambda y = 0,\\[6pt]
\mathcal{L}_\lambda = \dfrac{x^2}{4} + y^2 - 5 = 0.
\end{cases}
$$

Из первых двух уравнений:

$$
\lambda\,\frac{x}{2} \;=\; 2\lambda y \quad \Rightarrow \quad x = 4y,
$$

$$\text{(случай }\lambda=0\text{ невозможен, так как } e^{-(x+y)}\neq 0\text{).}$$


Подстановка в ограничение:

$$
\frac{(4y)^2}{4}+y^2=5 \;\Rightarrow\; 4y^2+y^2=5 \;\Rightarrow\; y^2=1 \;\Rightarrow\; y=\pm 1,\; x=\pm 4.
$$

Кандидаты на условные экстремумы

$$
(4,1)\quad \text{и}\quad (-4,-1).
$$

---

### Классификация (минимум/максимум)

$$
z \;=\; e^{-(x+y)} \;\; \text{монотонно убывает по сумме } (x+y).
$$

$$
\frac{x^2}{4} + y^2 = 5 \;\; \text{— уравнение эллипса}.
$$

Максимальное значение \(x+y\) на эллипсе достигается при \((4,1)\), минимальное — при \((-4,-1)\). Поэтому

$$
(4,1)\; \Rightarrow\; x+y=5 \; \Rightarrow\; z=e^{-5}\ \text{ — условный минимум};
$$

$$
(-4,-1)\; \Rightarrow\; x+y=-5 \; \Rightarrow\; z=e^{5}\ \text{ — условный максимум}.
$$

---

### Ответ

$$
\text{Условный минимум:}\quad (x,y)=(4,1),\quad z_{\min}=e^{-5}.
$$

$$
\text{Условный максимум:}\quad (x,y)=(-4,-1),\quad z_{\max}=e^{5}.
$$


### Задание 4. 

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import CategoricalNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# 1. Загрузка данных
data = pd.read_csv("exam.csv") # Загрузка данных из CSV файла
print(data.head())

# 2. Кодирование категориальных признаков
le_test = LabelEncoder() 
le_age = LabelEncoder() 
le_status = LabelEncoder()

data['Test'] = le_test.fit_transform(data['Test']) # Кодирование столбца 'Test'
data['Age_Group'] = le_age.fit_transform(data['Age_Group']) # Кодирование столбца 'Age_Group'
data['Status'] = le_status.fit_transform(data['Status']) # Кодирование целевого столбца 'Status'

# 3. Разделение данных на обучающую и тестовую выборки
X = data[['Test', 'Age_Group']] # Признаки
y = data['Status'] # Целевая переменная
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

# 4. Обучение модели Наивного Байеса
model = CategoricalNB() # Инициализация модели
model.fit(X_train, y_train) # Обучение модели

# 5. Предсказание и оценка качества модели
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred) # Точность
precision = precision_score(y_test, y_pred) # Точность положительных предсказаний
recall = recall_score(y_test, y_pred) # Полнота
f1 = f1_score(y_test, y_pred) # F1-мера

print("Accuracy:", round(accuracy, 3)) # Вывод результатов точности
print("Precision:", round(precision, 3)) # Вывод результатов точности положительных предсказаний
print("Recall:", round(recall, 3)) # Вывод результатов полноты
print("F1-score:", round(f1, 3)) # Вывод результатов F1-меры
```

Модель корректно классифицирует случаи с точностью 77%.

Показатель recall = 0.853 говорит о том, что алгоритм хорошо выявляет заболевших (низкое число пропущенных).

Небольшое снижение precision до 0.725 означает, что часть здоровых людей могла быть ошибочно отнесена к заболевшим.

Итоговая метрика F1 = 0.784 показывает хорошее качество классификации.

### Задание 5.

```python
# Метод Ньютона для оптимизации f(x) = x^4 + 3x^3 - 12x^2 + 7x - 2
# Старт: x0 = 10, точность: 1e-4

def f(x):
    return x**4 + 3*x**3 - 12*x**2 + 7*x - 2

def df(x):   # f'(x) первая производная
    return 4*x**3 + 9*x**2 - 24*x + 7

def d2f(x):  # f''(x) вторая производная
    return 12*x**2 + 18*x - 24

def newton_optimize(x0=10.0, tol=1e-4, max_iter=100): # Метод Ньютона
    x = float(x0) # начальная точка
    for it in range(1, max_iter + 1):
        g  = df(x) # градиент в текущей точке
        H  = d2f(x) # гессиан (вторая производная) в текущей точке
        if abs(H) < 1e-12:   # защита от деления на ~0 
            raise ZeroDivisionError("Вторая производная близка к нулю — метод Ньютона остановлен.")
        x_new = x - g / H # шаг метода Ньютона
        if abs(x_new - x) < tol or abs(g) < tol: # условие остановки
            x = x_new
            return x, f(x), it
        x = x_new
    return x, f(x), max_iter # возврат результата после max_iter

x_star, f_star, n_iter = newton_optimize(x0=10.0, tol=1e-4) # Запуск метода Ньютона

# Классификация (минимум/максимум) по знаку f''(x*)
kind = "минимум" if d2f(x_star) > 0 else ("максимум" if d2f(x_star) < 0 else "перелом/неопределено")

print(f"x* = {x_star:.6f}") # Вывод найденной точки x* где метод Ньютона остановился
print(f"f(x*) = {f_star:.6f}") # Вывод значения функции в точке x* 
print(f"итераций: {n_iter}") # Вывод количества итераций
print(f"тип точки: {kind}") # Вывод типа точки
```

```python
# Распределение: P(X=k) для k=0..6 заданы, P(X=7) неизвестна.
p = {0: 0.10, 1: 0.30, 2: 0.10, 3: 0.25, 4: 0.05, 5: 0.05, 6: 0.10}

# 1) Вероятность получить 7 писем
p7 = 1.0 - sum(p.values()) # P(X=7) = 1 - сумма остальных вероятностей
p[7] = p7  # добавим в распределение P(X=7)

# 2) Математическое ожидание E[X] 
EX = sum(k * pk for k, pk in p.items())# вычисление E[X]

# 3) Дисперсия Var[X] = E[X^2] - (E[X])^2
EX2 = sum((k * k) * pk for k, pk in p.items())
Var = EX2 - EX * EX 

# Вывод c округлением до сотых (требование задачи)
print(f"P(X=7) = {p7:.2f}") # Вывод вероятности P(X=7)
print(f"E[X] = {EX:.2f}") # Вывод математического ожидания E[X]
print(f"Var[X] = {Var:.2f}") # Вывод дисперсии Var[X]
```

```python
# Число звонков за час: N ~ Poisson(λ=2). Ищем P(2 ≤ N ≤ 4).
def fact(n):
    f = 1
    for i in range(2, n+1):
        f *= i
    return f

# простая экспонента 
import math
lam = 2.0
p = 0.0
for k in (2, 3, 4):
    p += math.exp(-lam) * (lam**k) / fact(k)

print("Искомая вероятность: P(2 ≤ N ≤ 4).")
print(f"Результат: P = {p:.3f}")
```
